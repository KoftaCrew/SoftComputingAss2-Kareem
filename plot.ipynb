```python
with open("curve_fitting_input.txt") as f:
    input = f.readlines()
    data = []

    # Reading data line by line
    t = int(input[0])
    current_line = 1
    for i in range(t):
        line = input[current_line].split()
        n = int(line[0])
        d = int(line[1])
        current_line += 1

        data_points = []

        for j in range(n):
            line = input[current_line].split()
            x = float(line[0])
            y = float(line[1])
            current_line += 1
            data_points.append((x, y))
        data.append(data_points)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Plot scatter plot
def plotScatter(data):
    x = [i[0] for i in data]
    y = [i[1] for i in data]
    plt.scatter(x, y, color='red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Scatter plot')
    plt.show()
```

```python
# Plot polynomial
def plotPolynomial(data, polynomial):
    x = [i[0] for i in data]
    y = [i[1] for i in data]
    plt.scatter(x, y, color='red')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polynomial plot')
    x = np.linspace(min(x), max(x), 100)
    y = np.polyval(polynomial, x)
    plt.plot(x, y)
    plt.show()
```

```python
plotPolynomial(data[0], [0.721209078332223, 0.6004645108888699, -0.4844235300639106, 0.06086362310032257][::-1])
```
